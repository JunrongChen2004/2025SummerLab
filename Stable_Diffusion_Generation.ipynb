```python
from diffusers import StableDiffusionPipeline
import torch

# Check if CUDA is available
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load the pre-trained Stable Diffusion model
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", torch_dtype=torch.float16 if device == "cuda" else torch.float32)
pipe = pipe.to(device)

# Function to generate image with specified number of apples
def generate_image_with_apples(number_of_apples, pipe):
    prompt = f"A realistic photo of {number_of_apples} apples on the floor, with natural lighting, high detail, and sharp focus"
    image = pipe(prompt).images[0]
    return image
image = generate_image_with_apples(1, pipe)
image.save(f"apple_{1}.png")
# Generate and save images
for i in range(2, 11):
    image = generate_image_with_apples(i, pipe)
    image.save(f"apple_{i}.png")

print("Generated 10 images with 1 to 10 apples.")
```
